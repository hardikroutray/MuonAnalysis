```python


tree_mubphi1 = RT.TChain('events')
tree_mubphi1.Add("~/nobackup/Jupyter/flat_dimuon_tree_BPhi_m0p5_ct1_flattree_vf.root")
tree_mubphi2 = RT.TChain('events')
tree_mubphi2.Add("~/nobackup/Jupyter/flat_dimuon_tree_BPhi_m2_ct10_flattree_vf.root")
tree_mubphi3 = RT.TChain('events')
tree_mubphi3.Add("~/nobackup/Jupyter/flat_dimuon_tree_BPhi_m4_ct100_flattree_vf.root")

```

```python
tree_muproofbphi1 = RT.TProofChain(tree_mubphi1, True)
tree_muproofbphi2 = RT.TProofChain(tree_mubphi2, True)
tree_muproofbphi3 = RT.TProofChain(tree_mubphi3, True)
```

```python
c1=RT.TCanvas('c1','c1',600,600)

h1b = RT.TH1F("h1b","",250,0,5)
h2b = RT.TH1F("h2b","",250,0,5)
h3b = RT.TH1F("h3b","",250,0,5)

h1bc = RT.TH1F("h1bc","",250,0,5)
h2bc = RT.TH1F("h2bc","",250,0,5)
h3bc = RT.TH1F("h3bc","",250,0,5)

tree_muproofbphi1.Draw('dimuon_mass>>h1bc','lxy > 0.0 && lxy < 0.2','')
tree_muproofbphi1.Draw('dimuon_mass_uncorr>>h1b','lxy > 0.0 && lxy < 0.2','')

tree_muproofbphi2.Draw('dimuon_mass>>h2bc','lxy > 0.0 && lxy < 0.2','')
tree_muproofbphi2.Draw('dimuon_mass_uncorr>>h2b','lxy > 0.0 && lxy < 0.2','')


tree_muproofbphi3.Draw('dimuon_mass>>h3bc','lxy > 0.0 && lxy < 0.2','')
tree_muproofbphi3.Draw('dimuon_mass_uncorr>>h3b','lxy > 0.0 && lxy < 0.2','')


GoodPlotFormat(h1b, "thickline", 91, 1)
GoodPlotFormat(h1bc, "thickline", 86, 1)

GoodPlotFormat(h2b, "thickline", 94, 1)
GoodPlotFormat(h2bc, "thickline", 8, 1)

GoodPlotFormat(h3b, "thickline", 98, 1)
GoodPlotFormat(h3bc, "thickline", 3, 1)


h1b.Scale(1/h1b.Integral())
h1bc.Scale(1/h1bc.Integral())

h2b.Scale(1/h2b.Integral())
h2bc.Scale(1/h2bc.Integral())

h3b.Scale(1/h3b.Integral())
h3bc.Scale(1/h3bc.Integral())


h1b.GetYaxis().SetRangeUser(0,1.0)
h1b.GetXaxis().SetRangeUser(0,100)
h1b.GetYaxis().SetTitle("Fraction of Events")
h1b.GetXaxis().SetTitle("Dimuon Mass(in GeV)")
h1b.GetYaxis().SetTitleOffset(1.5)
RT.gStyle.SetLabelColor(1)


h1b.Draw("HIST")
h1bc.Draw("HIST SAME")

h2b.Draw("HIST SAME")
h2bc.Draw("HIST SAME")

h3b.Draw("HIST SAME")
h3bc.Draw("HIST SAME")

legend = RT.TLegend(0.2, 0.6, 0.9, 0.8)  
legend.SetBorderSize(0)                                                                                                                                                              
legend.SetFillColorAlpha(0,0)                                                                                                                                                               
legend.SetTextSize(0.03) 
# legend.SetHeader("0 cm < lxy < 11 cm")

legend.AddEntry(h1b , "B->Phi(mass = 0.5 GeV, ctau = 1 mm) Uncorrected", "F")

legend.AddEntry(h2b , "B->Phi(mass = 2 GeV, ctau = 10 mm) Uncorrected", "F")

legend.AddEntry(h3b , "B->Phi(mass = 4 GeV, ctau = 100 mm) Uncorrected", "F")   

legend.AddEntry(h1bc , "B->Phi(mass = 0.5 GeV, ctau = 1 mm) Corrected", "F")

legend.AddEntry(h2bc , "B->Phi(mass = 2 GeV, ctau = 10 mm) Corrected", "F")                                                                                                           

legend.AddEntry(h3bc , "B->Phi(mass = 4 GeV, ctau = 100 mm) Corrected", "F")   


legend.SetTextFont(22)
legend.Draw()


#AddCMSLumi(c1, "", "#font[12]{Preliminary}" )
c1.Draw()

t = RT.TText(0.45,0.93,"0.0 cm < lxy < 0.2 cm")
t.SetTextAlign(22)
t.SetTextColor(8)
t.SetTextFont(42)
t.SetTextSize(20)
t.SetTextAngle(0)
t.SetNDC(1)
t.Draw()
```
