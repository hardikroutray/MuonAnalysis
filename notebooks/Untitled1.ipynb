```python
tree_muhzd1 = RT.TChain('events')
tree_muhzd1.Add("~/nobackup/Jupyter/flat_dimuon_tree_HZdZd_m2_ct100_flattree_vf.root")
tree_muhzd2 = RT.TChain('events')
tree_muhzd2.Add("~/nobackup/Jupyter/flat_dimuon_tree_HZdZd_m8_ct10_flattree_vf.root")
tree_muhzd3 = RT.TChain('events')
tree_muhzd3.Add("~/nobackup/Jupyter/flat_dimuon_tree_HZdZd_m15_ct1_flattree_vf.root")
```

```python
c1=RT.TCanvas('c1','c1',600,600)

h1zd = RT.TH1F("h1zd","",200,0,20)
h2zd = RT.TH1F("h2zd","",200,0,20)
h3zd = RT.TH1F("h3zd","",200,0,20)

h1zdc = RT.TH1F("h1zdc","",200,0,20)
h2zdc = RT.TH1F("h2zdc","",200,0,20)
h3zdc = RT.TH1F("h3zdc","",200,0,20)

tree_muproofhzd1.Draw('dimuon_mass>>h1zdc','lxy > 0.0 && lxy < 0.2','')
tree_muproofhzd1.Draw('dimuon_mass_uncorr>>h1zd','lxy > 0.0 && lxy < 0.2','')

tree_muproofhzd2.Draw('dimuon_mass>>h2zdc','lxy > 0.0 && lxy < 0.2','')
tree_muproofhzd2.Draw('dimuon_mass_uncorr>>h2zd','lxy > 0.0 && lxy < 0.2','')

tree_muproofhzd3.Draw('dimuon_mass>>h3zdc','lxy > 0.0 && lxy < 0.2','')
tree_muproofhzd3.Draw('dimuon_mass_uncorr>>h3zd','lxy > 0.0 && lxy < 0.2','')


GoodPlotFormat(h1zd, "thickline", 71, 1)
GoodPlotFormat(h1zdc, "thickline", 86, 1)

GoodPlotFormat(h2zd, "thickline", 65, 1)
GoodPlotFormat(h2zdc, "thickline", 8, 1)

GoodPlotFormat(h3zd, "thickline", 51, 1)
GoodPlotFormat(h3zdc, "thickline", 3, 1)

h1zd.Scale(1/h1zd.Integral())
h1zdc.Scale(1/h1zdc.Integral())

h2zd.Scale(1/h2zd.Integral())
h2zdc.Scale(1/h2zdc.Integral())

h3zd.Scale(1/h3zd.Integral())
h3zdc.Scale(1/h3zdc.Integral())

h1zd.GetYaxis().SetRangeUser(0,1.0)
h1zd.GetXaxis().SetRangeUser(0,100)
h1zd.GetYaxis().SetTitle("Fraction of Events")
h1zd.GetXaxis().SetTitle("Dimuon Mass(in GeV)")
h1zd.GetYaxis().SetTitleOffset(1.5)
RT.gStyle.SetLabelColor(1)


h1zd.Draw("HIST")
h1zdc.Draw("HIST SAME")

h2zd.Draw("HIST SAME")
h2zdc.Draw("HIST SAME")

h3zd.Draw("HIST SAME")
h3zdc.Draw("HIST SAME")

legend = RT.TLegend(0.2, 0.6, 0.9, 0.8)  
legend.SetBorderSize(0)                                                                                                                                                              
legend.SetFillColorAlpha(0,0)                                                                                                                                                               
legend.SetTextSize(0.03) 
# legend.SetHeader("0 cm < lxy < 11 cm")

legend.AddEntry(h1zd , "H->ZdZd(mass = 2 GeV, ctau = 100 mm) Uncorrected", "F")
legend.AddEntry(h2zd , "H->ZdZd(mass = 8 GeV, ctau = 10 mm) Uncorrected", "F")                                                                                                           
legend.AddEntry(h3zd , "H->ZdZd(mass = 15 GeV, ctau = 1 mm) Uncorrected", "F")                                                                                                           

legend.AddEntry(h1zdc , "H->ZdZd(mass = 2 GeV, ctau = 100 mm) Corrected", "F")
legend.AddEntry(h2zdc , "H->ZdZd(mass = 8 GeV, ctau = 10 mm) Corrected", "F")                                                                                                           
legend.AddEntry(h3zdc , "H->ZdZd(mass = 15 GeV, ctau = 1 mm) Corrected", "F")                                                                                                           


legend.SetTextFont(22)
legend.Draw()


AddCMSLumi(c1, 10.1, "#font[12]{Preliminary}" )
c1.Draw()

t = RT.TText(0.45,0.93,"0.0 cm < lxy < 0.2 cm")
t.SetTextAlign(22)
t.SetTextColor(8)
t.SetTextFont(42)
t.SetTextSize(20)
t.SetTextAngle(0)
t.SetNDC(1)
t.Draw()
```
